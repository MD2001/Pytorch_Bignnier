## 0. Computer vision libaries in PyTorch

* [`torchvision`] (https://www.learnpytorch.io/03_pytorch_computer_vision/) - base
domain library for PyTorch computer vision
* `torchvision.datasets` - get datasets and data loading functions for computer
vision here
* `torchvision.models` - get pretrained computer vision models that you can
leverage for your own problems
* `torchvision.transforms` - functions for manipulating your vision data (images)
to be suitable for use with an ML model
* `torch.utils.data.Dataset` - Base dataset class for PyTorch.
* `torch. utils.data.DataLoader` - Creates a Python iterable over a datas

```python
# Import PyTorch
import torch
from torch import nn

# Import torchvision
import torchvision
from torchvision import datasets
from torchvision import transforms
from torchvision.transforms import ToTensor

# Import matplotlib for visualization
import matplotlib.pyplot as plt
```

# 1.Import data set

```python
Training_Data=datasets.FashionMNIST(
    root='data',
    train=True,
    download=True,
    transform=ToTensor(),
    target_transform=None
)

Test_Data=datasets.FashionMNIST(
    root='data',
    train=False,
    download=True,
    transform=ToTensor(),
    target_transform=None
)
print(f"the input data shape for this data set is {Training_Data.data.shape}\n ex: {Training_Data.data[0,0]} ")
print(f"the target data shape for this data set is {Training_Data.targets.shape} ex:{Training_Data.targets[0]}")
```

```python
PicNumber= 55
plt.imshow(Training_Data.data[PicNumber],cmap='gray')
plt.title(f"{PicNumber}- The {Training_Data.classes[Training_Data.targets[PicNumber]]}")
plt.axis(False);      # the ';' disable any unrelvant data that colab show
```

```python
#the data loader that make mini-batchs to use in training
from torch.utils.data import DataLoader
BATCH_SIZE=32
Trining_batch=DataLoader(dataset =Training_Data,
                         batch_size =BATCH_SIZE,
                         shuffle=True)
Test_batch=DataLoader(dataset =Test_Data,
                         batch_size =BATCH_SIZE,
                         shuffle=False)
Trining_batch,Test_batch
```

```python
# Let's check out what what we've created
print(f"DataLoaders: {Trining_batch, Test_batch}")
print(f"Length of train_dataloader: {len(Trining_batch) } batches of {BATCH_SIZE} ... ")
print(f"Length of test_dataloader: {len(Test_batch) } batches of {BATCH_SIZE} ... ")
```
